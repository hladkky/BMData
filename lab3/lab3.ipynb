```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode

%matplotlib inline
sns.set(style="darkgrid")
pal = sns.palplot(sns.color_palette("Blues"));
pal2 = sns.palplot(sns.dark_palette("purple"));
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
sns.distplot(planets['distance'], kde=False, bins=100)
```

```python
sns.jointplot(x='orbital_period', y='mass', data=planets, kind='kde',
              color='b')
```

```python
sns.pairplot(planets, hue = 'number')
```

```python
sns.barplot(x='method', y='year', data=planets, estimator=np.std)
```

```python
sns.boxplot(x='number', y='year', data=planets, palette="Set1")
```

```python
sns.stripplot(x='number', y='method', data=planets)
```

```python
sns.heatmap(planets.corr())
```

```python
g = sns.FacetGrid(planets, col='number', row='method')
g = g.map(plt.hist, 'year')
```

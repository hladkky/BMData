```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y = x^2")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'palegoldenrod')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'lightseagreen')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('y=x^2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0,0.5,0.5,0.5])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('y=x^2')

axes2.plot(y,x)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('y=x^2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, 'silver')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(y,x, 'darkmagenta')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
```

```python
fig, axes = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**3, label='x^3')
axes.plot(x, x**(1/2), label='sqrt(x)')
axes.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi = 200)
```

```python
fig, axes = plt.subplots()
axes.plot(x, pow(x, 1/2), marker='o', markersize = 5, markerfacecolor='magenta', lw = 2, ls = ':')
```

```python
fig.savefig("fig.pdf")
```
